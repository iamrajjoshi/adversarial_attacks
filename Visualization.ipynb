# Visualization

```python
%matplotlib inline

import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
epsilons = [0, .05, .1, .15, .2, .25, .3]
```

```python
fgsm_accuracies_imagenette = torch.load("results/fgsm_accuracies_imagenette.pt")
fgsm_examples_imagenette = torch.load("results/fgsm_examples_imagenette.pt")

igs_accuracies_imagenette = torch.load("results/igs_accuracies_imagenette.pt")
igs_examples_imagenette = torch.load("results/igs_examples_imagenette.pt")

llc_accuracies_imagenette = torch.load("results/llc_accuracies_imagenette.pt")
llc_examples_imagenette = torch.load("results/llc_examples_imagenette.pt")
```

```python
plt.figure(figsize=(6,6))
plt.plot(epsilons, fgsm_accuracies_imagenette, "*-", label="FGSM")
plt.plot(epsilons, igs_accuracies_imagenette, "*-", label="Iterative FGSM")
plt.plot(epsilons, llc_accuracies_imagenette, "*-", label="LLCM")
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, .35, step=0.05))
plt.title("Accuracy vs Epsilon (Imagenette)")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
fgsm_accuracies_mnist = torch.load("results/fgsm_accuracies_mnist.pt")
fgsm_examples_mnist = torch.load("results/fgsm_examples_mnist.pt")

igs_accuracies_mnist = torch.load("results/igs_accuracies_mnist.pt")
igs_examples_mnist = torch.load("results/igs_examples_mnist.pt")

llc_accuracies_mnist = torch.load("results/llc_accuracies_mnist.pt")
llc_examples_mnist = torch.load("results/llc_examples_mnist.pt")
```

```python
plt.figure(figsize=(6,6))
plt.plot(epsilons, fgsm_accuracies_mnist, "*-", label="FGSM")
plt.plot(epsilons, igs_accuracies_mnist, "*-", label="Iterative FGSM")
plt.plot(epsilons, llc_accuracies_mnist, "*-", label="LLCM")
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, .35, step=0.05))
plt.title("Accuracy vs Epsilon (MNIST)")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
imagenette_labels = ["tench", "English springer", "cassette player", "chainsaw", "church", "French horn", "garbage truck", "gas pump", "golf ball", "parachute"]
```

```python
orig,adv,orig_ex,adv_ex = fgsm_examples_imagenette[1][2]

plt.figure(figsize=(8,16))

plt.subplot(1,2,1)
plt.xticks([], [])
plt.yticks([], [])


plt.title("{}".format(imagenette_labels[orig]))
orig_ex = orig_ex.transpose(1,2,0)
plt.imshow(orig_ex)

plt.subplot(1,2,2)
plt.xticks([], [])
plt.yticks([], [])

plt.title("{}".format(imagenette_labels[adv]))
adv_ex = adv_ex.transpose(1,2,0)
plt.imshow(adv_ex)

plt.tight_layout()
plt.show()
```
